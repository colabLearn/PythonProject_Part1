```python
# Nested loop:
for i in range (1,13):
    if i == 1:
        print("   ", end=' ')
    else:
        print("{:3d}".format(i), end=' ')
print()
for j in range(1,13):
    for k in range(1, 13):
        print("{:3d}".format(j*k), end=' ')
    print()
```
